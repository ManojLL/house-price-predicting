```python
import pandas as pd

# read the csv file (data set) using pandas
data = pd.read_csv('ml_house_data_set_updated.csv')
data.head(5)
```

```python
# cleaning the loaded data set 
# remove unwanted columns 
del data['house_number']
del data['street_name']
del data['unit_number']
del data['zip_code']
```

```python

feature_data = pd.get_dummies(data, columns=['garage_type','city'])
feature_data.head(5)
```

```python
# remove house price from feature_data
del feature_data['sale_price']
```

```python
from sklearn.model_selection import train_test_split
X = feature_data
Y = data['sale_price']

# split data into traing & testing sets
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size =0.3,random_state=0)

```

```python
#fit model
from sklearn import ensemble
from sklearn.metrics import mean_absolute_error
import joblib

model = ensemble.GradientBoostingRegressor(
    n_estimators=1000,
    learning_rate=0.3,
    max_depth=7,
    min_samples_leaf=9,
    max_features=0.3,
    loss='huber',
    random_state=0
)
model.fit(X_train,Y_train)
```

```python
mse = mean_absolute_error(Y_train, model.predict(X_train))
print("Training Set Mean Absolute Error: %.4f" % mse)
```

```python
mse = mean_absolute_error(Y_test, model.predict(X_test))
print("Test Set Mean Absolute Error: %.4f" % mse)
```
